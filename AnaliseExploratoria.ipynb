```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lendo o arquivo CSV
```

```python
dados = pd.read_csv('dados.csv', sep=';')
```

```python
print(dados.head())
```

```python
print(dados[['PIB', 'VALOREMPENHO']].describe())
```

```python
# Criando indicador de empenho proporcional ao PIB
```

```python
dados['EMPENHO_POR_PIB'] = dados['VALOREMPENHO'] / dados['PIB']
```

```python
# Visualização: Distribuição do PIB
```

```python
plt.figure(figsize=(10,6))
sns.histplot(dados['PIB'], kde=True, bins=20, color='skyblue')
plt.title('Distribuição do PIB dos Municípios')
plt.xlabel('PIB')
plt.ylabel('Frequência')
plt.show()
```

```python
# Visualização: Distribuição do Valor Empenhado
```

```python
plt.figure(figsize=(10,6))
sns.histplot(dados['VALOREMPENHO'], kde=True, bins=20, color='salmon')
plt.title('Distribuição do Valor Empenhado em Diárias')
plt.xlabel('Valor Empenhado')
plt.ylabel('Frequência')
plt.show()
```

```python
# Gráfico de dispersão: PIB vs. Empenho
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=dados, x='PIB', y='VALOREMPENHO')
plt.title('Relação entre PIB e Valor Empenhado')
plt.xlabel('PIB')
plt.ylabel('Valor Empenhado')
plt.show()
```

```python
# Top 10 municípios com maior empenho proporcional ao PIB
```

```python
top10_proporcao = dados.sort_values('EMPENHO_POR_PIB', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10_proporcao, x='MUNICIPIO', y='EMPENHO_POR_PIB', palette='viridis')
plt.title('Top 10 Municípios com Maior Empenho Proporcional ao PIB')
plt.xticks(rotation=45)
plt.ylabel('Empenho / PIB')
plt.show()
```
